# 2. Предобработка

```python
missing = df.isna().sum().sort_values(ascending=False)
missing[missing > 0]
```

```python
print("Количество полных дубликатов:", df.duplicated().sum())
```

```python
df["calendar_date"] = pd.to_datetime(df["calendar_date"], errors="coerce")

print("Пустых дат после преобразования:", df["calendar_date"].isna().sum())
```

```python
df = df.sort_values(["SEQN", "calendar_date"]).reset_index(drop=True)
df.head()
```

```python
print("Число наблюдений:", len(df))
print("Число уникальных участников:", df["SEQN"].nunique())
print("Среднее число дней на участника:", round(len(df) / df["SEQN"].nunique(), 2))
```
